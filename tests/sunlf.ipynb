```python
import networkx as nx
import matplotlib.pyplot as plt

# 1. 创建一个空的无向图
G = nx.Graph()

# 2. 添加节点（可以带属性）
G.add_node("A", color="red", size=10)
G.add_node("B", color="blue", size=20)
G.add_nodes_from([
    ("C", {"color": "green", "size": 15}),
    ("D", {"color": "yellow"})
])

# 3. 添加边（可以带权重）
G.add_edge("A", "B", weight=1.5)
G.add_edge("B", "C", weight=2.0)
G.add_edges_from([
    ("A", "D", {"weight": 0.5}),
    ("C", "D")
])

# 4. 基本查询
print("节点数量:", G.number_of_nodes())  # 4
print("边数量:", G.number_of_edges())    # 4
print("A的邻居:", list(G.neighbors("A"))) # ['B', 'D']

# 5. 可视化
pos = nx.spring_layout(G)  # 布局算法
nx.draw(G, pos, with_labels=True, node_color=[G.nodes[n]['color'] for n in G.nodes])
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.show()

# 6. 保存为GraphML文件（与你代码中的格式一致）
nx.write_graphml(G, "example.graphml")

```
